# Crittografia per Sostituzione

Prima di tutto definisco la funzione per creare l'associazione tra i caratteri del nostro alfabeto e i caratteri dell'alfabeto crittografato.

```python
# Funzione per creare il dizionario di sostituzione
def crea_alfabeto_sostituzione():
    alfabeto = 'abcdefghijklmnopqrstuvwxyz'
    alfabeto_sostituito = 'cbafedihglkjmnorqptsruvzwyx' # SCRIVI QUI IL TUO ALFABETO SOSTITUTIVO
    sostituzione = dict(zip(alfabeto, alfabeto_sostituito))
    return sostituzione
```

Creiamo ora le funzioni per criptare e decriptare il testo

```python
# Funzione per crittografare il testo
def crittografa_testo(testo, sostituzione):
    testo_cifrato = ''

    # SCRIVI UN CICLO PER SOSTITUIRE TUTTE LE LETTERE DEL TESTO CON IL NUOVO ALFABETO
    for carattere in testo:
      isUpper = carattere.isupper()
        #Se è una lettera minuscola
      if carattere.lower() in sostituzione:
            if isUpper:
                testo_cifrato += sostituzione[carattere.lower()].upper()
            else:
                testo_cifrato += sostituzione[carattere.lower()]
      else:
            testo_cifrato += carattere

    return testo_cifrato

# Funzione per decrittografare il testo
def decrittografa_testo(testo, sostituzione):
    # Creiamo il dizionario inverso per la decrittografia
    sostituzione_inversa = {v: k for k, v in sostituzione.items()}

    testo_decifrato = ''

    # SCRIVI UN CICLO PER SOSTITUIRE TUTTE LE LETTERE DEL TESTO CRIPTATO CON IL VERO ALFABETO

    for carattere in testo:
        #Se è una lettera minuscola
        if carattere.lower() in sostituzione_inversa:
            if carattere.islower():
                testo_decifrato += sostituzione_inversa[carattere] #contrario
            else:
                testo_decifrato += sostituzione_inversa[carattere.lower()].upper() #aggiunto lower
        else:
            testo_decifrato += carattere

    return testo_decifrato
```

## Test del nostro metodo crittografico

```python
# Esempio di utilizzo
testo_originale = "Ciao, come stai?"

# Creiamo l'alfabeto di sostituzione
sostituzione = crea_alfabeto_sostituzione()

print("Testo originale:", testo_originale)

# Criptiamo il testo
testo_cifrato = crittografa_testo(testo_originale, sostituzione)
print("Testo cifrato:", testo_cifrato)

# Decriptiamo il testo
testo_decifrato = decrittografa_testo(testo_cifrato, sostituzione)
print("Testo decifrato:", testo_decifrato)
```
